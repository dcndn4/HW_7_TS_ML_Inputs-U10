# Linear Regression Error and Accuracy Metrics

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
%matplotlib inline
```

# Import Data

```python
from pathlib import Path

csvpath="C:/Users/CS_Knit_tinK_SC/Documents/My Data Sources/111321/03_weather_data_austin_2010.csv"

#csvpath = Path('../Resources/weather_data_austin_2010.csv')
df = pd.read_csv(csvpath, parse_dates=True, index_col='Date')
df.head()
```

```python
df = df.loc['2010 May':'2010 Jul']
df.head()
```

```python
df['Temperature'].plot()
```

# Set Features (X) and Target (y)

In this example, we want to use the week of the year to predict the temperature

```python
# Create a DataFrame of the input features (X)
# Export the Temperature column as a DataFrame
X = df['Temperature'].to_frame()  # simpler, results in just a series

# = df.Temperature.values.reshape(-1,1)  # more complex

print(f"X has {X.shape[0]} rows and {X.shape[1]} column of data.")
X.head()
```

```python
# Generate a column with the week of the year
X['Week_of_Year'] = X.index.weekofyear
X.head(52)
```

<font color='red'>Comment</font>

drop first = https://stackoverflow.com/questions/50176096/removing-redundant-columns-when-using-get-dummies

Week of year is not a continuous number span, but is rather categorical set of data now.

Here, each week is given its own week in the equation. We have many variable/coefficients now. This is **multiple regression.**

```python
# Binary encode the weekly column to use as new input features for the model
X_binary_encoded = pd.get_dummies(X, columns=['Week_of_Year']) #, drop_first = True)
X_binary_encoded.tail()
#X_binary_encoded.columns
```

<font color='red'>Comment</font>

On dataframe copy: https://stackoverflow.com/questions/27673231/why-should-i-make-a-copy-of-a-data-frame-in-pandas

and :https://community.dataquest.io/t/equals-vs-shallow-copy-vs-deep-copy-in-pandas-dataframes/514515

.copy() method is used to copy the df and ensure no unwanted changes are made to the original data. Here, it doesn't seem to matter though.

```python
# The target for our model is to predict the Temperature
#y = df['Temperature'].copy()
y = df['Temperature']
```

```python
# The target Temperature column must be deleted from the input features
X_binary_encoded = X_binary_encoded.drop('Temperature', axis=1) 
# = .drop(columns=['Temperature']) this does same thing, might be clearer
X_binary_encoded.head()
```

# Linear Regression

```python
# Create the Linear Regression model
model = LinearRegression()

# Fit the model to the data (Train the model)
model.fit(X_binary_encoded, y)

# Predict the temperature using the week of the year as inputs
predictions = model.predict(X_binary_encoded)
```

# Evaluating the Model

<font color='red'>Comment</font>

r-square is fairly low.

```python
import numpy as np
from sklearn.metrics import mean_squared_error, r2_score

# The default scoring method is the R2 score
score = model.score(X_binary_encoded, y)
r2 = r2_score(y, predictions)

print(f"Score: {score}, r2: {r2}")
```

```python
# Mean Squared Error
mse = mean_squared_error(y, predictions)
mse
```

```python
# Root Mean Squared Error
rmse = np.sqrt(mse)
rmse
```

```python
# Standard deviation of the temperature
np.std(y)
```

# Plotting the Regression Line
