# Autocorrelation and Partial Autocorrelation Analysis

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
plot_acf(df, lags=20)
```

<font color='red'>Comment</font>

Looks like month of year determines flight count. It is seasonal! A bit of a jump around the new year.

Below, we see ARMA(2, 2) would be appropriate to capture the first 2 lags. Seasonal ARIMA/SARIMA (https://machinelearningmastery.com/sarima-for-time-series-forecasting-in-python/) might be applied here.

```python
plot_pacf(df, zero=False, lags=20)
```

<font color='red'>Comment</font>

So, in the above, we wouldn't want to use ARMA, because it assumes stationarity.

If we wanted to use ARMA, we'd have to stationarize it, using .pct_change(), .diff(), or another approach.

We can skip that manual data processing step by jumping straight to ARIMA (Auto Regressive Integrated Moving Average).

The 'I' stationarizes the data for us.

ARIMA(2, 0, 2) = ARMA(2, 2)

The middle number is the order of differencing. If the time series needs only one .diff() in order to become stationary, use 1. If passengers needed to be differenced once, and then a second time, it you would use 2.

# ARIMA Model

```python
from statsmodels.tsa.arima_model import ARIMA
model = ARIMA(df['NumberPassengers'], order=(2, 1, 2))
```

```python
results = model.fit()
```

```python
results.summary()
```

<font color='red'>Comment</font>

Keep proceeding to end before interpreting these metrics!!!

```python
# Visualize the forecast
pd.DataFrame(results.forecast(steps=3)[0]).plot(title="Passenger Travel Forecast")
```

```python
# A second model with a different order
model2 = ARIMA(df['NumberPassengers'], order=(2, 1, 4))
```

```python
res2 = model2.fit()
```

```python
res2.summary()
```

<font color='red'>Comment</font>

The second model above was more complex (2, 1, 4).

The model below was less complex (2, 1, 2).

AIC/BIC penalizes for former model due to higher order complexity. Each additional lag adds a variable to the model.


More complex model:

1. AIC	1338.585

2. BIC	1362.288

Simpler model:

1. AIC	1344.043

2. BIC	1361.821

Lower is better, so favor the simpler model.

So...use ACF/PACF and AIC/BIC to determine your model orders.

More about ARIMA [here](https://people.duke.edu/~rnau/arimrule.htm).
