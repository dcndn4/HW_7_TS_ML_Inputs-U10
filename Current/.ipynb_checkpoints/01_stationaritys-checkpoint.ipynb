```python
import numpy as np
import pandas as pd
import random
%matplotlib inline
```

```python
# Data Generation
# Warning! Boilerplate code
n = 100
x = np.arange(n)  # Return evenly spaced values within a given interval.
y_stationary = np.random.normal(0, 1, n)  # Draw random samples from a normal (Gaussian) distribution.
y_nonstationary = [random.random() + (x * .01) for x in range(100)]

df = pd.DataFrame({
    "Stationary": y_stationary,
    "Nonstationary": y_nonstationary
})
```

<font color='red'>Comment</font>

Stationary data is constant mean and variance.

```python
df.Stationary.plot()
```

<font color='red'>Comment</font>

Non stationary data because mean and variance change over time.

```python
df.Nonstationary.plot()
```

<font color='red'>Comment</font>

ADF tells you if the series is stationary. You need to unpack the contents, and refer to the docs to understand what the function returns.

The p-value is the second value in the output. A p-value less than 0.05 (not 0.5) tells that the series is stationary.

The more negative the ADF Statistic is, the more likely it is stationary.

```python
# Apply the Augmented Dickey-Fuller test to determine if the above is stationary
from statsmodels.tsa.stattools import adfuller

# Store the results of the test in the variable result
result = adfuller(df.Stationary)

# Access the contents of the results:
print('ADF Statistic: %f' % result[0])
print('p-value:        %f ' % result[1])
print('Lags used:      %d' % result[2])
print('Critical Values:')
for key, value in result[4].items():
    print((key, value))
```

```python
# Store the results of the test in the variable result
result = adfuller(df.Nonstationary)
# Access the contents of the results:
print('ADF Statistic: %f' % result[0])
print('p-value:        %f ' % result[1])
print('Lags used:      %d' % result[2])
print('Critical Values:')
for key, value in result[4].items():
    print((key, value))
```

# Non-Stationary to Stationary Techniques

### Technique 1: Applying a Percent Change

```python
# Convert to stationary data by taking the `pct_change` of the non-stationary data.
df['Returns'] = df.Nonstationary.pct_change()
```

```python
# Drop the NaNs created by the pct_change function
df = df.dropna()
```

```python
# Plot the returns as a line chart. 
# You should no longer see the upward trend in the data
df.Returns.plot()
```

### Technique 2: Applying a Diff

```python
# Subtract the upward trend for the non-stationary data using the `diff()` function
df['Diff'] = df.Nonstationary.diff()
```

```python
# Drop the NaNs created from the diff function
df = df.dropna()
```

```python
# Plot the 
df.Diff.plot()
```

<font color='red'>Comment</font>
both .diff() and .pct_change() make data stationary, aka "stationarizing".

.pct_change() essentially normalizes the data (the denominator does the scaling)

If the data is exponential, take the logarithm.

See `/Activities/diff_vs_pct_change.xlsx`

There are a variety of ways to make data stationary...
