```python
# Focus on the 2008-2009 Financial Crisis Period
returns = sp500.loc['2008':'2009'].Close.pct_change() * 100  # times 100 for easier viewing!
returns = returns.dropna()
returns.tail()
```

<font color='red'>Comment</font>

Creating GARCH is similar to creating ARIMA model...

# GARCH

```python
import arch
from arch import arch_model
```

```python
# 'p' and 'q'  are akin to the 'p' and 'q' of an ARMA model.
# 'vol="GARCH"' means that we're using a GARCH model.
# The 'mean="Zero"' means that we're estimating a GARCH.
model = arch_model(returns, mean="Zero", vol="GARCH", p=1, q=1)

# p = AR(1), q = MA(1)
```

```python
# Fit the GARCH Model
res = model.fit(disp="off")

# The arguments iter and disp are used in fit() to suppress estimation output.
```

```python
# Summarize the model results
res.summary()
```

```python
# Plot the model estimate of annualized volatility
fig = res.plot(annualize='D')

# https://arch.readthedocs.io/en/latest/univariate/generated/generated/arch.univariate.base.ARCHModelResult.plot.html
# String containing frequency of data that indicates plot should contain annualized volatility. 
# Supported values are ‘D’ (daily), ‘W’ (weekly) and ‘M’ (monthly), which scale variance by 252, 52, and 12, respectively.
```

<font color='red'>Comment</font>

With ARIMA, we were estimating returns.

With GARCH, we are estimating volatility. Below, we forecast 3 days ahead.

# Forecasting from GARCH

```python
# Construct Volatility Forecasts for the next 3 days
forecast_horizon = 3

# Take the last day of the data we used above. 
# If forecast horizon is 3, then the resulting 'h.1', 'h.2', and 'h.3' 
# are the forecasts for the following 3 days.
forecasts = res.forecast(start='2009-12-31', horizon=forecast_horizon, reindex=True)
forecasts
```

```python
# Annualize the forecast
intermediate = np.sqrt(forecasts.variance.dropna() * 252)
intermediate
```

# Visualizing the forecast

```python
# The name of the column here is the date of the forecast.
# Each row represents the forecast of volatility for the following days.
# Transposing makes the forecast easier to plot
final = intermediate.dropna().T
final
```

```python
final.plot(title='GARCH volatility forecast')
```
